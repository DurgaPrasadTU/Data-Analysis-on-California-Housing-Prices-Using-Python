```python
#To calculate the median values by creating a user defined function
import pandas as pd                          #importing the pandas dataframe
def median_value(housing,columns):           # creating 'median_value function name and assigning the arguments
  medians = housing[columns].median()        # calculate the medians for the specified columns in the housing DataFrame
  return medians

select_columns = ['latitude',                          # selecting all the requried columns to calculate the median values.
                  'longitude',
                  'housing_median_age',
                  'total_rooms',
                  'total_bedrooms',
                  'population',
                  'households',
                  'median_income',
                  'median_house_value',
                  'ocean_proximity']

medians = median_value(housing,select_columns)           # calling the function
print(medians)                                        # print the median of the selected columns

#From below result we get to know the median of latitude is 34.26
# median of longitude is -118.4900
# median of housing_median_age is 29.0
# median of total_rooms is 2127.0
# median of total_bedrooms is 435.00
# median of population is 1166.0
# median of households is 409.0
# median of median_income is 3.5348
# median of median_house_value is 179700.0
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
import numpy as np                #importing numpy library
conditions = [
              (housing['total_bedrooms'] <=10),
              (housing['total_bedrooms'] >=11)&(housing['total_bedrooms'] <=1000),      #setting condition for different total_bedroom_size
              (housing['total_bedrooms'] > 1000)
             ]
values = ['small','medium','large']                   # creating list for the values

housing['total_bedroom_size']=np.select(conditions,values)            # using numpy library and selecting the condition and values

housing
```
